```python
#@title Word Compare Function
# compares letters in a with letters in b. 
# ex, if a == *el*o & b == hello, return True; if a == *el*o & b == matza, return False
# word_compare("happy","ha***") returns True
def word_compare(a,b):
  if len(a) != len(b):
    return False
  for x in range(len(a)):
    if a[x] == b[x] or b[x] == "*":
      continue
    else:
      return False
  return True
```
